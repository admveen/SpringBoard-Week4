**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# I'm just changing the label on the first row -- when we transpose this will turn into a column label corresponding to the borough/region codes
properties.iloc[0,0] = 'Borough Code' 

data_raw = properties.T
data_raw = data_raw.reset_index()

data_raw.columns = data_raw.iloc[0,:]
data_raw.drop(0, inplace = True)
data_raw.head()
```

```python
data_raw.columns
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
data_raw = data_raw.rename(columns = {'Unnamed: 0':'Borough Name'})
data_raw.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
data_raw = pd.melt(data_raw, id_vars = ['Borough Name', 'Borough Code'], var_name = 'Date', value_name = 'Average Price')
data_raw.head()
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

I'm checking to see if there are any more null values. There are none. So we will move forwards with final restucturing of the data.

```python
data_raw.isna().any()
```

```python
# enforcing float type on the monthly average price
BoroughPriceTimeSeries = data_raw.set_index(['Borough Name','Date']).sort_index()
BoroughPriceTimeSeries['Average Price'] = BoroughPriceTimeSeries['Average Price'].astype('float')
```

```python
BoroughPriceTimeSeries.head()
```

```python
BoroughPriceTimeSeries.index.levels[0]
```

```python
# I'm downsampling (via mean aggregation) this data into a dataframe that displays average price per year for each district.
YearlySeries = BoroughPriceTimeSeries.groupby([pd.Grouper(level='Borough Name'),  
            pd.Grouper(level='Date', freq='A')]
          ).mean()
YearlySeries.head()
```

```python
#pandas seems to be having trouble doing selection on partial datetimes (it's fine on slicing) so I've gone ahead and reindexed with just the year only
locale_index = YearlySeries.index.levels[0]
year_index = YearlySeries.index.levels[1].year
reindexer = pd.MultiIndex.from_product([locale_index, year_index], names = ['Borough Name', 'Year'])
HousingPrices = YearlySeries.set_index(reindexer)
HousingPrices.head()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
# Create a new dataframe HousingGrowth that just looks at the averages for 1998 and 2018 and we unstack the dataframe by Year level.
idx = pd.IndexSlice
HousingGrowth = HousingPrices.loc[idx[:, [1998,2018]], :]

HousingGrowth = HousingGrowth.unstack(level = 'Year')
HousingGrowth.head()
```

```python
HousingGrowth.loc[:, ('Average Price', 'Growth')] = HousingGrowth['Average Price'][2018] / HousingGrowth['Average Price'][1998]
```

```python
HousingGrowth.head()
```

The following now gets the boroughs/regions ranked by their growth from 1998 to 2018. Growth here has been ranked by the ratio of the yealy housing prices averages of 2018 to 1998. Note that the uppercase only labeled 'boroughs' seem to actually be national regions as opposed to London boroughs. Before filtering these out its good to note that the London greater metroplitan area's average is higher than the rest of the country -- perhaps not surprisingly. Also, there is the row 'England' which is the national housing average.

```python
HousingGrowth = HousingGrowth.sort_values(('Average Price','Growth'), ascending = False )
HousingGrowth
```
